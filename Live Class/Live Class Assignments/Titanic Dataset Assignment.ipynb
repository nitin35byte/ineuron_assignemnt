### 10.From a  cabin column seperate text and numeric value . 

```python
char=[]
num=[]
for i in df['Cabin']:
    string = ''
    numeric = ''
    if type(i)==str:
        for j in i:
            if j>='A' and j <='Z' or  j>='a' and j<='z':
                string +=j
            else:
                numeric +=j
            char.append(string)
            num.append(numeric)
        else:
            char.append(i)
            num.append(i)
```
